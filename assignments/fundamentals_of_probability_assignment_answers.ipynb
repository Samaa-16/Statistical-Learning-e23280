# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


```python
import numpy as np

# Given parameters
radius_disk = 10
unit_from_boundary = 1

# --- Part 1: Model 1 (Uniformly distributed with respect to area) ---
# The event A is that the point lies within 1 unit of the boundary.
# This means the distance from the center is greater than (radius_disk - unit_from_boundary).
# So, the point is in an annulus with outer radius R and inner radius R - 1.

# Total area of the disk
area_total = np.pi * (radius_disk**2)

# Inner radius for the annulus (points further than 1 unit from boundary)
inner_radius_annulus = radius_disk - unit_from_boundary

# Area of the annulus (event A)
area_event_A = np.pi * (radius_disk**2 - inner_radius_annulus**2)

# Probability for Model 1
p_A_model1 = area_event_A / area_total

print(f"--- Model 1: Uniformly distributed with respect to area ---")
print(f"Total disk radius: {radius_disk}")
print(f"Inner radius for event A: {inner_radius_annulus}")
print(f"Area of total disk: {area_total:.2f} sq units")
print(f"Area of annulus (event A): {area_event_A:.2f} sq units")
print(f"P(A) (Model 1): {p_A_model1:.4f}\n")

# --- Part 2: Model 2 (Distance from center uniformly distributed) ---
# The distance to the point from the center (let's call it 'r') is chosen uniformly from [0, 10].
# The event A is that the point lies within 1 unit of the boundary, meaning r > (radius_disk - unit_from_boundary).
# So, r must be in the interval (9, 10].

# Total range for radial distance
radial_range_total = radius_disk - 0

# Range for event A (radial distance between 9 and 10)
radial_range_event_A = radius_disk - inner_radius_annulus

# Probability for Model 2
p_A_model2 = radial_range_event_A / radial_range_total

print(f"--- Model 2: Distance from center uniformly distributed ---")
print(f"Total radial range: [0, {radius_disk}]")
print(f"Radial range for event A: ({inner_radius_annulus}, {radius_disk}]")
print(f"Length of total radial range: {radial_range_total}")
print(f"Length of radial range for event A: {radial_range_event_A}")
print(f"P(A) (Model 2): {p_A_model2:.4f}\n")

# --- Part 3: Explanation of why the two answers differ ---
print("--- Explanation of the difference ---")
print("The two answers differ because the underlying probability models define 'randomness' differently:")
print("\n1.  **Model 1 (Uniformly distributed with respect to area):**")
print("    In this model, any given unit of area within the disk has an equal probability of containing the point. For a disk, infinitesimal area elements near the boundary cover a much larger region than infinitesimal area elements near the center. Specifically, the area of an annulus increases with its distance from the center. Therefore, points are implicitly more likely to be found further from the center. The event 'within 1 unit of the boundary' (i.e., radius between 9 and 10) constitutes 19% of the total area, hence P(A) = 0.19.")
print("\n2.  **Model 2 (Distance from center uniformly distributed):**")
print("    In this model, the radial distance 'r' from the center is chosen uniformly, meaning any interval of radial distances of the same length has an equal probability. For example, the interval [0,1] has the same probability (1/10) as the interval [9,10]. This implies that there is an equal chance of finding the point at any specific distance from the center. Since the event 'within 1 unit of the boundary' corresponds to the radial distance interval (9,10], which is 1 unit long out of a total radial range of 10 units, its probability is 1/10 = 0.10.")
print("\n**Conclusion:** Model 1 gives a higher probability for event A because it inherently weights the outer regions of the disk more heavily due to their larger area, whereas Model 2 gives equal weight to all radial distances, making the outermost unit interval of radius less probable when considering the overall area.")
```

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

```python
import numpy as np

# Let R be the radius of the circle. For simplicity, we can set R=1.
R = 1

# The side length of an equilateral triangle inscribed in a circle of radius R
# is s = R * sqrt(3).
# A chord's length L exceeds s if L > R * sqrt(3).
# The length of a chord L is related to its distance 'd' from the center by L = 2 * sqrt(R^2 - d^2).
# So, L > R * sqrt(3) implies 2 * sqrt(R^2 - d^2) > R * sqrt(3).
# Squaring both sides: 4 * (R^2 - d^2) > 3 * R^2
# 4R^2 - 4d^2 > 3R^2
# R^2 > 4d^2
# d^2 < R^2 / 4
# d < R / 2.
# Thus, the condition for the chord length to exceed the side of the inscribed equilateral triangle is that
# its distance from the center 'd' must be less than R/2.

print(f"For a circle of radius R={R}:\n")
print(f"The side length of an inscribed equilateral triangle is s = {R * np.sqrt(3):.2f}\n")
print(f"A chord's length exceeds this value if its distance 'd' from the center is less than R/2 = {R/2:.2f}.\n")

# --- Model 1: Random Endpoints ---
print("--- Model 1: Choosing two random points on the circumference ---")
# Choose two points randomly and independently on the circumference.
# Let the angular separation of these two points be 2*alpha, where alpha is the angle
# formed by the radius to an endpoint and the chord's perpendicular bisector.
# The chord length L = 2R * sin(alpha).
# We need L > R * sqrt(3) => 2R * sin(alpha) > R * sqrt(3) => sin(alpha) > sqrt(3)/2.
# This means alpha must be in the range (pi/3, 2pi/3).
# However, alpha is typically defined as the angle from 0 to pi/2, for the half-chord.
# More directly: The angle subtended by the chord at the center is theta. L = 2R sin(theta/2).
# We need 2R sin(theta/2) > R sqrt(3) => sin(theta/2) > sqrt(3)/2.
# This implies theta/2 must be between pi/3 and 2pi/3 (or 60 to 120 degrees).
# Since theta is uniformly distributed between 0 and 2pi, and we consider the shorter arc,
# the angle theta can be considered from 0 to pi. For the chord length, we are interested in theta from 0 to pi.
# If theta is the angle between the two points, then theta can range from 0 to pi.
# We need theta/2 > pi/3, so theta > 2pi/3.
# The range of theta for L > R*sqrt(3) is (2pi/3, pi]. The length of this range is pi - 2pi/3 = pi/3.
# The total possible range for theta (for distinct chord lengths) is (0, pi].
# So the probability is (pi/3) / pi = 1/3.
p_model1 = 1/3
print(f"Probability for Model 1: {p_model1:.4f}\n")

# --- Model 2: Random Midpoint ---
print("--- Model 2: Choosing a random midpoint within the circle ---")
# Choose a point uniformly at random within the circle. This point is the midpoint of the chord.
# We need the distance 'd' of this midpoint from the center to be less than R/2.
# The probability is the ratio of the area of a smaller circle (with radius R/2)
# to the area of the full circle (with radius R).
# Area_small_circle = pi * (R/2)^2 = pi * R^2 / 4.
# Area_full_circle = pi * R^2.
# Probability = (pi * R^2 / 4) / (pi * R^2) = 1/4.
p_model2 = 1/4
print(f"Probability for Model 2: {p_model2:.4f}\n")

# --- Model 3: Random Radius and Random Point on Radius ---
print("--- Model 3: Choosing a random radius and a random point on it ---")
# Choose a radius of the circle uniformly at random.
# Choose a point on this radius uniformly at random (distance 'd' from the center, 0 <= d <= R).
# This point defines the midpoint of a chord perpendicular to the chosen radius.
# We need the distance 'd' to be less than R/2.
# Since 'd' is chosen uniformly from [0, R], the probability that d < R/2 is (R/2) / R = 1/2.
p_model3 = 1/2
print(f"Probability for Model 3: {p_model3:.4f}\n")

# --- Explanation of why the answers differ ---
print("--- Explanation of the Differences (Bertrand's Paradox) ---")
print("Bertrand's Paradox arises because the phrase 'at random' is ambiguous when applied to geometric objects, and different mathematically 'natural' interpretations lead to different probability distributions for the chord's properties.")
print("\n1.  **Model 1 (Random Endpoints):** This model implicitly favors longer chords. Chords are distributed such that their lengths are more likely to be closer to the radius than to the diameter or very short. The distribution of the chord's distance from the center is not uniform.")
print("\n2.  **Model 2 (Random Midpoint):** This model implicitly favors shorter chords. By choosing the midpoint uniformly within the circle, points closer to the center are chosen with higher probability than points further away (if you consider the distance from the center 'd' as a random variable). The area density of midpoints is uniform, but this means that the distribution of 'd' is not uniform; there are more midpoints near the center. Consequently, chords with midpoints near the center (which are longer) are less likely.")
print("\n3.  **Model 3 (Random Radius and Point on Radius):** This model yields a uniform distribution for the distance 'd' from the center to the chord. Any distance 'd' between 0 and R is equally likely. Therefore, the probability that 'd' is less than R/2 is simply the ratio of the favorable range to the total range.")
print("\n**Conclusion:** Each method sets up a different sample space or a different probability measure on the space of chords, leading to distinct results. The paradox highlights the importance of precisely defining the underlying probability distribution when dealing with continuous random variables in geometry.")
```

#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



```python
# Given probabilities for the number of emergency transformer failures (X)
P_X_pmf = {
    0: 0.50,
    1: 0.30,
    2: 0.15,
    3: 0.05
}

print("--- Q3) Transformer Failures ---\n")

# 1. Construct a probability space (Ω, F, P)
print("1. Probability Space (Ω, F, P):")
omega = set(P_X_pmf.keys())
print(f"   Ω (Sample Space) = {omega} (possible number of failures)")
print(f"   F (Sigma-algebra) = P(Ω) (the power set of Ω, as Ω is finite)")
print(f"   P (Probability Measure): Defined by P(X=x) as given:")
for x, p in P_X_pmf.items():
    print(f"     P({{X={x}}}) = {p}")
print("\n")

# 2. Define X as a random variable on this space
print("2. Definition of X as a Random Variable:")
print("   X: Ω → ℝ, where X(ω) = ω for ω ∈ Ω.")
print("   That is, the random variable X directly represents the number of failures.")
print("\n")

# 3. Determine the probability mass function (PMF) of X
print("3. Probability Mass Function (PMF) of X:")
print("   The PMF is given by the problem statement:")
for x, p in P_X_pmf.items():
    print(f"     P(X={x}) = {p}")
print("\n")

# 4. Compute the expectation E[X]
E_X = sum(x * p for x, p in P_X_pmf.items())
print(f"4. Expectation E[X]:")
print(f"   E[X] = Σ [x * P(X=x)] = (0*0.50) + (1*0.30) + (2*0.15) + (3*0.05)")
print(f"   E[X] = {E_X:.2f}\n")

# 5. Compute the variance Var(X)
E_X2 = sum((x**2) * p for x, p in P_X_pmf.items())
Var_X = E_X2 - (E_X**2)
print(f"5. Variance Var(X):")
print(f"   E[X^2] = Σ [x^2 * P(X=x)] = (0^2*0.50) + (1^2*0.30) + (2^2*0.15) + (3^2*0.05)")
print(f"   E[X^2] = {E_X2:.2f}")
print(f"   Var(X) = E[X^2] - (E[X])^2 = {E_X2:.2f} - ({E_X:.2f})^2")
print(f"   Var(X) = {Var_X:.2f}\n")

# 6. Find P(X >= 1) and P(X >= 2)
P_X_ge_1 = sum(P_X_pmf[x] for x in P_X_pmf if x >= 1)
# Alternatively: P_X_ge_1 = 1 - P_X_pmf[0]

P_X_ge_2 = sum(P_X_pmf[x] for x in P_X_pmf if x >= 2)

print(f"6. Probabilities P(X ≥ 1) and P(X ≥ 2):")
print(f"   P(X ≥ 1) = P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = {P_X_ge_1:.2f}")
print(f"   P(X ≥ 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = {P_X_ge_2:.2f}\n")

# 7. Briefly explain the practical meaning of the expectation and variance
print("7. Practical Meaning of Expectation and Variance:")
print(f"   **Expectation (E[X] = {E_X:.2f}):** This represents the average number of transformer failures expected per day in this district over a long period. On average, the district can expect {E_X:.2f} failures daily.")
print(f"   **Variance (Var(X) = {Var_X:.2f}):** This measures the spread or variability in the number of daily failures around the expected value. A variance of {Var_X:.2f} indicates how much the actual number of failures on any given day is likely to deviate from the average of {E_X:.2f}. A higher variance would imply a greater unpredictability in the daily failure count.")
```

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


```python
import numpy as np
from scipy.stats import beta
from scipy.special import beta as beta_func_constant

print("--- Q4) Probability of Defective Items ---\n")

# Parameters for the Beta distribution
alpha = 2
beta_param = 18 # Renamed to avoid conflict with scipy.stats.beta

# 1. Construct a probability space (Ω, F, P)
print("1. Probability Space (Ω, F, P):")
print(f"   Ω (Sample Space) = [0, 1] (the interval for proportions)")
print(f"   F (Sigma-algebra) = B([0, 1]) (the Borel σ-algebra on [0,1])")
print(f"   P (Probability Measure): Induced by the probability density function f(x)")
print("\n")

# 2. Define X as a continuous random variable on this space
print("2. Definition of X as a Continuous Random Variable:")
print(f"   X: Ω → ℝ, where X(ω) = ω.")
print(f"   X maps each possible proportion of defective items to itself.")
print("\n")

# 3. Write down the probability density function of X
print("3. Probability Density Function (PDF) of X:")
beta_constant = beta_func_constant(alpha, beta_param)
print(f"   f(x) = (1 / B({alpha},{beta_param})) * x^({alpha-1}) * (1-x)^({beta_param-1}), for 0 < x < 1")
print(f"   where B({alpha},{beta_param}) is the Beta function, approximately {beta_constant:.6f}.")
print("\n")

# Define the PDF function for calculations (optional, but good for understanding)
def f(x, a, b):
    if 0 < x < 1:
        return (x**(a-1) * (1-x)**(b-1)) / beta_func_constant(a, b)
    return 0

# 4. Verify that f is a valid probability density function
print("4. Verification of the PDF:")
print("   a. Non-negativity: For 0 < x < 1, x^(α-1) and (1-x)^(β-1) are non-negative, and B(α,β) is positive.")
print("      Thus, f(x) ≥ 0 for all x.")
print("   b. Integration to 1: By definition of the Beta function, the integral of x^(α-1)(1-x)^(β-1) from 0 to 1 is B(α,β).")
print("      Therefore, ∫[0,1] f(x) dx = ∫[0,1] (1 / B(α,β)) * x^(α-1) * (1-x)^(β-1) dx = (1 / B(α,β)) * B(α,β) = 1.")
print("      The PDF is valid.")
print("\n")

# 5. Compute the expectation E[X]
E_X = alpha / (alpha + beta_param)
print(f"5. Expectation E[X]:")
print(f"   E[X] = α / (α + β) = {alpha} / ({alpha} + {beta_param}) = {E_X:.4f}")
print("\n")

# 6. Compute the variance Var(X)
Var_X = (alpha * beta_param) / (((alpha + beta_param)**2) * (alpha + beta_param + 1))
print(f"6. Variance Var(X):")
print(f"   Var(X) = (αβ) / ((α + β)^2 * (α + β + 1)) = ({alpha} * {beta_param}) / (({alpha} + {beta_param})^2 * ({alpha} + {beta_param} + 1))")
print(f"   Var(X) = {Var_X:.6f}")
print("\n")

# 7. Find the probability that the defective proportion is at most 0.10
P_X_le_0_10 = beta.cdf(0.10, alpha, beta_param)
print(f"7. P(X ≤ 0.10):")
print(f"   P(X ≤ 0.10) = ∫[0, 0.10] f(x) dx")
print(f"   Using the CDF for Beta distribution, P(X ≤ 0.10) = {P_X_le_0_10:.4f}")
print("\n")

# 8. Find the probability that the defective proportion lies between 0.05 and 0.15
P_0_05_le_X_le_0_15 = beta.cdf(0.15, alpha, beta_param) - beta.cdf(0.05, alpha, beta_param)
print(f"8. P(0.05 ≤ X ≤ 0.15):")
print(f"   P(0.05 ≤ X ≤ 0.15) = ∫[0.05, 0.15] f(x) dx")
print(f"   Using the CDF for Beta distribution, P(0.05 ≤ X ≤ 0.15) = {P_0_05_le_X_le_0_15:.4f}")
print("\n")

# 9. Briefly explain why a Beta distribution is a reasonable model for this situation
print("9. Why a Beta distribution is a reasonable model:")
print("   The Beta distribution is ideal for modeling random variables that represent probabilities or proportions, as its support is naturally bounded between 0 and 1. In this context, the proportion of defective items is a value between 0% and 100%.")
print("   Its shape parameters (α and β) allow for great flexibility in modeling various scenarios, such as symmetric, skewed, or U-shaped distributions. Here, with α=2 and β=18, the distribution is heavily right-skewed, indicating that a low proportion of defective items is most likely, which is desirable in manufacturing.")
print("\n")

# 10. Interpret the expectation and variance in the context of production quality
print("10. Interpretation of Expectation and Variance:")
print(f"    **Expectation (E[X] = {E_X:.4f}):** This value represents the average or mean proportion of defective items expected in a large batch over time. An expected value of {E_X:.2%} suggests that, on average, 10% of items produced are defective. This is a key indicator of the overall quality level of the manufacturing process.")
print(f"    **Variance (Var(X) = {Var_X:.6f}):** This measures the spread or variability of the defective proportion around the expected value. A variance of {Var_X:.6f} indicates how consistent the manufacturing process is. A low variance, as seen here, suggests that the proportion of defective items tends to stay close to the average, implying a stable and predictable production quality. A higher variance would mean more fluctuations, with some days having very few defectives and others having many, making quality control more challenging.")
```

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


```python
import numpy as np
from scipy.stats import triang

print("--- Q5) Expected Project Completion Time ---\n")

# Given parameters for the Triangular distribution
a = 4   # Shortest possible time (minimum)
b = 10  # Longest possible time (maximum)
c = 6   # Most likely completion time (mode)

# 1. Specify the support of X
print("1. Support of X:")
print(f"   The support of X is the interval [a, b], which is [{a}, {b}] hours.\n")

# 2. Write down the probability density function f_X(x)
print("2. Probability Density Function f_X(x):")
print(f"   f_X(x) = ")
print(f"     {{ (2 * (x - a)) / ((b - a) * (c - a)),   if a <= x < c  (i.e., {a} <= x < {c}) ")
print(f"     {{ (2 * (b - x)) / ((b - a) * (b - c)),   if c <= x <= b (i.e., {c} <= x <= {b}) ")
print(f"     {{ 0,                                   otherwise ")
print("\n")

# 3. Verify that f_X integrates to 1 over its support
print("3. Verification that f_X integrates to 1:")
print("   The PDF of a triangular distribution forms a triangle with base (b-a) and height 2/(b-a) at the mode c. ")
print("   The area of this triangle is (1/2) * base * height = (1/2) * (b - a) * (2 / (b - a)) = 1. ")
print("   Thus, f_X integrates to 1 over its support.\n")

# 4. Compute the expectation E[X]
E_X = (a + b + c) / 3
print(f"4. Expectation E[X]:")
print(f"   E[X] = (a + b + c) / 3 = ({a} + {b} + {c}) / 3 = {E_X:.2f} hours\n")

# 5. Compute the variance Var(X)
Var_X = (a**2 + b**2 + c**2 - a*b - a*c - b*c) / 18 # Corrected formula for variance
print(f"5. Variance Var(X):")
print(f"   Var(X) = (a^2 + b^2 + c^2 - ab - ac - bc) / 18")
print(f"   Var(X) = ({a**2} + {b**2} + {c**2} - {a*b} - {a*c} - {b*c}) / 18")
print(f"   Var(X) = {Var_X:.2f}\n")

# 6. Interpret the values of a, b, and c
print("6. Interpretation of a, b, and c:")
print(f"   - **a = {a} hours:** Represents the **optimistic (shortest) time** estimate for the installation. Under the most ideal conditions, the task cannot be completed faster than this.")
print(f"   - **b = {b} hours:** Represents the **pessimistic (longest) time** estimate for the installation. Even with all foreseeable problems, the task should not take longer than this.")
print(f"   - **c = {c} hours:** Represents the **most likely time** for the installation. This is the duration that is expected to occur most frequently.\n")

# 7. Explain briefly why a Triangular distribution is a reasonable model
print("7. Why a Triangular distribution is a reasonable model:")
print("   The Triangular distribution is often used in project management when historical data is scarce, but expert judgment can provide minimum, maximum, and most likely estimates. It's simple, intuitive, and captures the uncertainty within a bounded range. Its shape reflects that values around the 'most likely' point are indeed more probable, and probabilities decrease linearly towards the extremes (min and max). This fits well for tasks like construction where experts can provide these three key estimates.\n")

# For scipy.stats.triang, the parameters are (c-a)/(b-a) for the mode and loc=a, scale=b-a
mode_param = (c - a) / (b - a)

# 8. Compute P(X <= 5)
P_X_le_5 = triang.cdf(5, mode_param, loc=a, scale=b-a)
print(f"8. P(X ≤ 5):")
print(f"   P(X ≤ 5) = {P_X_le_5:.4f}\n")

# 9. Compute P(5 <= X <= 8)
P_X_le_8 = triang.cdf(8, mode_param, loc=a, scale=b-a)
P_5_le_X_le_8 = P_X_le_8 - P_X_le_5
print(f"9. P(5 ≤ X ≤ 8):")
print(f"   P(5 ≤ X ≤ 8) = {P_5_le_X_le_8:.4f}\n")

# 10. Briefly explain how this model could be used in project planning or sensitivity analysis
print("10. Project Planning and Sensitivity Analysis Use:")
print("    In project planning, this model provides a probabilistic estimate of task duration rather than a single fixed value. It allows project managers to:")
print("    - **Estimate Project Completion Probabilities:** By combining triangular distributions for multiple tasks (e.g., using Monte Carlo simulation), one can estimate the probability of completing an entire project by a certain deadline.")
print("    - **Perform Sensitivity Analysis:** Project managers can vary the 'a', 'b', and 'c' parameters to understand how changes in estimates impact overall project timelines and costs. This helps identify critical tasks with high variability that might need closer monitoring or contingency planning.")
print("    - **Set Realistic Deadlines:** Instead of just using the most likely time, the expectation (E[X]) and probabilities (like P(X ≤ x)) help in setting more realistic and defensible deadlines and in communicating risks to stakeholders.")
```

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Compute the expectation $E[X]$ using the PERT formula

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


```python
import numpy as np
from scipy.stats import beta

# Given parameters for PERT distribution
a = 8   # optimistic
b = 20  # pessimistic
c = 11  # most likely

print("--- Q6) Expected Project Completion Time Using PERT ---\n")

# 1. Compute the values of the shape parameters alpha and beta
alpha = 1 + 4 * ((c - a) / (b - a))
beta_param = 1 + 4 * ((b - c) / (b - a))
print(f"1. Shape Parameters:\n   alpha = {alpha:.4f}\n   beta = {beta_param:.4f}\n")

# 2. Specify the support of X
print(f"2. Support of X: [{a}, {b}] days\n")

# 3. Write down the PDF
print("3. Probability Density Function f_X(x):")
print(f"   f(x) = [(x-{a})^(alpha-1) * ({b}-x)^(beta-1)] / [B(alpha, beta) * ({b-a})^(alpha+beta-1)]")
print("   for 8 <= x <= 20.\n")

# 4. State the probability space
print("4. Probability Space (̉Ή, F, P):")
print(f"   ̉Ή = [{a}, {b}] (The interval of possible days)")
print("   F = Borel ̃΃-algebra on [8, 20]")
print("   P = Probability measure induced by the PERT density function.\n")

# 5. Define X as a random variable
print("5. Random Variable X: ̉Ή -> R where X(̹̉) = ̹̉ (the duration in days).\n")

# 6. Compute Expectation E[X]
E_X_pert = (a + 4*c + b) / 6
print(f"6. Expectation E[X] (PERT formula):\n   E[X] = (a + 4c + b) / 6 = ({a} + 4*{c} + {b}) / 6 = {E_X_pert:.4f} days\n")

# 7. Compute Variance Var(X)
Var_X_pert = ((E_X_pert - a) * (b - E_X_pert)) / 7
print(f"7. Variance Var(X) (Approximation):\n   Var(X) = [(E[X]-a)(b-E[X])] / 7 = {Var_X_pert:.4f}\n")

# 8. Compute Standard Deviation
sigma_pert = np.sqrt(Var_X_pert)
print(f"8. Standard Deviation ̳̃: {sigma_pert:.4f} days\n")

# 9. Comparison: Triangular vs PERT
E_X_tri = (a + b + c) / 3
print(f"9. Comparison:\n   Triangular E[X] = {E_X_tri:.4f}\n   PERT E[X] = {E_X_pert:.4f}")
print("   The PERT distribution is often more appropriate because it is 'smoother' and places less probability weight at the extremes compared to the Triangular distribution. It also gives the mode (most likely value) four times the weight of the extremes in the expectation calculation, making it less sensitive to extreme pessimistic or optimistic outliers.")

# 10. Interpretation
print("\n10. Interpretation:")
print(f"   - a={a}, b={20}, c={11}: The range and peak of our expert duration estimates.")
print(f"   - E[X]={E_X_pert:.2f}: On average, we expect the task to take about {E_X_pert:.2f} days.")
print(f"   - ̳̃={sigma_pert:.2f}: This represents the risk or uncertainty; approximately 68% of the time, the task will finish within +/- {sigma_pert:.2f} days of the expectation.")
```

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

```python
import numpy as np

# Given parameters for Q7
a = 5
c = 11

print("--- Q7) PERT Expectation Matching Intuition ---\n")

# 1. Determine the value of b such that E[X] = c
# Formula: E[X] = (a + 4c + b) / 6
# Set E[X] = c => c = (a + 4c + b) / 6
# 6c = a + 4c + b
# b = 2c - a
b = 2 * c - a

print(f"1. To have E[X] = c = {c} with a = {a}:")
print(f"   b = 2c - a = 2({c}) - {a} = {b} days.\n")

# 2. What happens to the variance?
# E[X] is now equal to c, so E[X] = 11.
E_X = c
Var_X = ((E_X - a) * (b - E_X)) / 7

print(f"2. Variance Calculation:")
print(f"   Var(X) = [(E[X] - a)(b - E[X])] / 7")
print(f"   Var(X) = [({E_X} - {a})({b} - {E_X})] / 7")
print(f"   Var(X) = [({E_X - a})({b - E_X})] / 7 = {Var_X:.4f}")
print("\nObservation: When the distribution is symmetric (c is exactly in the middle of a and b), the expectation equals the mode.")
```

#Q8) Probability of Being a Girl?

A family has two children. Assume each child is equally likely to be a boy or a girl, independently of the other.

What is the probability that both children are girls, given that **at least one** of them is a girl?

```python
import itertools

# Define the sample space for two children (B for Boy, G for Girl)
children = ['B', 'G']
sample_space = list(itertools.product(children, repeat=2))

print(f"Sample Space: {sample_space}")

# Event B: At least one is a girl
event_B = [outcome for outcome in sample_space if 'G' in outcome]

# Event A: Both are girls
event_A = [outcome for outcome in sample_space if outcome == ('G', 'G')]

# P(A | B) = Count(A and B) / Count(B)
# Since A is a subset of B in this case:
prob_both_girls_given_at_least_one = len(event_A) / len(event_B)

print(f"\n1. Outcomes with at least one girl: {event_B} (Count: {len(event_B)})")
print(f"2. Outcomes with both girls: {event_A} (Count: {len(event_A)})")
print(f"\nProbability P(Both Girls | At Least One Girl) = {len(event_A)}/{len(event_B)} = {prob_both_girls_given_at_least_one:.4f}")
```

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

Assume the sample space is

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

with probabilities

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

Define two discrete random variables on the same probability space:

* $X$ records only whether the item is good or defective:
  $$
  X(G_1)=0,\quad X(G_2)=0,\quad X(D_1)=1,\quad X(D_2)=1.

  $$
  Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4.

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that $\sigma(X)\subseteq \sigma(Y)$, and explain why $Y$ contains more information than $X$.

5. Find the marginal distribution of $X$.

6. Find the marginal distribution of $Y$.

7. Compute the joint probabilities
   $$
   P(X=0,Y=1),\quad P(X=0,Y=2),\quad P(X=1,Y=3),\quad P(X=1,Y=4).
   $$

8. Compute the conditional probabilities
   $$
   P(X=1\mid Y=3),\qquad P(X=1\mid Y=4),\qquad P(Y=3\mid X=1),\qquad P(Y=4\mid X=1).
   $$

9. Interpret the difference between conditioning on $X$ and conditioning on $Y$.

11. Give a Python based visualization of the model.

12. Find the marginal, joint, and conditional Shannon Entropy of $X$ and $Y$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given probabilities
probs = {
    'G1': 0.50,
    'G2': 0.20,
    'D1': 0.10,
    'D2': 0.20
}

print("--- Q9) Information Available in a Discrete Random Variable ---\n")

# 1. Probability Space
print("1. Probability Space (̉Ή, F, P):")
print("   ̉Ή = {G1, G2, D1, D2}")
print("   F = {∅, {G1}, {G2}, {D1}, {D2}, {G1,G2}, ..., ̉Ή} (Power set, 16 elements)")
print(f"   P defined by: P(G1)={probs['G1']}, P(G2)={probs['G2']}, P(D1)={probs['D1']}, P(D2)={probs['D2']}\n")

# 2 & 3. Sigma-algebras
print("2. ̃΃(X): Generated by X (Quality only)")
print("   X^-1({0}) = {G1, G2}, X^-1({1}) = {D1, D2}")
print("   ̃΃(X) = {∅, {G1, G2}, {D1, D2}, ̉Ή}\n")

print("3. ̃΃(Y): Generated by Y (Quality and Machine)")
print("   Since Y is distinct for every outcome, ̃΃(Y) = F (all 16 subsets).\n")

# 4. Comparison
print("4. Information Comparison:")
print("   ̃΃(X) ⊆ ̃΃(Y) is true because every set in ̃΃(X) is a union of sets in ̃΃(Y).")
print("   Y contains more information because knowing Y tells you the specific machine, whereas X does not.\n")

# 5 & 6. Marginal Distributions
p_X0 = probs['G1'] + probs['G2']
p_X1 = probs['D1'] + probs['D2']
print(f"5. Marginal X (0=Good, 1=Defective): P(X=0)={p_X0:.2f}, P(X=1)={p_X1:.2f}")
print(f"6. Marginal Y: P(Y=1)=0.5, P(Y=2)=0.2, P(Y=3)=0.1, P(Y=4)=0.2\n")

# 7 & 8. Joint and Conditional
print("7. Compute the joint probabilities")
print(f"   P(X=0, Y=1) = P({{G1}}) = {probs['G1']:.2f}")
print(f"   P(X=0, Y=2) = P({{G2}}) = {probs['G2']:.2f}")
print(f"   P(X=1, Y=3) = P({{D1}}) = {probs['D1']:.2f}")
print(f"   P(X=1, Y=4) = P({{D2}}) = {probs['D2']:.2f}\n")

print("8. Conditional Probabilities:")
print(f"   P(X=1|Y=3) = P(X=1, Y=3) / P(Y=3) = P({{D1}}) / P({{D1}}) = {probs['D1'] / probs['D1']:.0f} (since Y=3 means D1)\n")
print(f"   P(X=1|Y=4) = P(X=1, Y=4) / P(Y=4) = P({{D2}}) / P({{D2}}) = {probs['D2'] / probs['D2']:.0f} (since Y=4 means D2)\n")

print(f"   P(Y=3|X=1) = P(X=1, Y=3) / P(X=1) = P({{D1}}) / P(X=1) = {probs['D1'] / p_X1:.4f}")
print(f"   P(Y=4|X=1) = P(X=1, Y=4) / P(X=1) = P({{D2}}) / P(X=1) = {probs['D2'] / p_X1:.4f}\n")

# 9. Interpret the difference between conditioning on X and conditioning on Y
print("9. Interpret the difference between conditioning on X and conditioning on Y:")
print("   - Conditioning on X (quality) means we know if the item is good or defective. We still don't know which machine produced it if it's good (G1 or G2) or defective (D1 or D2).")
print("   - Conditioning on Y (quality and machine) means we know the exact outcome (e.g., G1, G2, D1, or D2). This provides full information. For example, if we know Y=3 (D1), then we know X=1 (defective) with certainty.\n")

# 10. Explain what information is available in sigma(X) but not in sigma(Y), and vice-versa
print("10. Explain information in ̃΃(X) vs. ̃΃(Y):")
print("    - Information available in ̃΃(X) but NOT in ̃΃(Y): None. Since ̃΃(X) ⊆ ̃΃(Y), all information in ̃΃(X) is also present in ̃΃(Y).")
print("    - Information available in ̃΃(Y) but NOT in ̃΃(X): The specific machine (Machine 1 or Machine 2) an item came from. For example, ̃΃(Y) distinguishes between {G1} and {G2}, while ̃΃(X) only contains their union {G1, G2}.\n")

# 11. Give a Python based visualization of the model.
print("11. Python-based Visualization of the Model:")
plt.figure(figsize=(7, 4))
plt.bar(['G1', 'G2', 'D1', 'D2'], list(probs.values()), color=['green', 'lightgreen', 'red', 'salmon'])
plt.title("Outcome Probabilities (̉Ή)")
plt.ylabel("Probability")
plt.show()

# 12. Find the marginal, joint, and conditional Shannon Entropy of X and Y.
print("12. Shannon Entropy Analysis:")
def entropy(p_list):
    return -sum(p * np.log2(p) for p in p_list if p > 0)

H_X = entropy([p_X0, p_X1])
H_Y = entropy(list(probs.values())) # H(Y) is the entropy of the 4 outcomes
H_XY = H_Y # Since Y uniquely determines the outcome, the joint entropy H(X,Y) = H(Y)

# Conditional Entropy H(Y|X) = H(X,Y) - H(X)
H_Y_given_X = H_XY - H_X

# Conditional Entropy H(X|Y) = 0 because Y provides full information about X
H_X_given_Y = 0

print(f"    H(X) = {H_X:.4f} bits (Uncertainty about quality: good/defective)")
print(f"    H(Y) = {H_Y:.4f} bits (Total uncertainty in outcomes: G1, G2, D1, D2)")
print(f"    H(X|Y) = {H_X_given_Y:.4f} bits (Uncertainty about X given Y is known: 0, as Y fully determines X)")
print(f"    H(Y|X) = {H_Y_given_X:.4f} bits (Uncertainty about Y given X is known: remaining uncertainty about machine given quality)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

print("--- Q10) Information in Continuous Random Variables ---\n")

# Parameters
# T ~ N(0, 1), P ~ N(0, 4)
mean = [0, 0]
cov = [[1, 0], [0, 4]] # Covariance matrix for independent T and P

# 1. Probability Space (Ω, F, P)
print("1. Probability Space (Ω, F, P):")
print("   Ω = ℝ² (The space of all possible (T,P) pairs).")
print("   F = ℬ(ℝ²) (The Borel σ-algebra on ℝ²).")
print("   P is the probability measure induced by the joint density f_T,P(t,p)).\n")

# 2. σ-algebra generated by X, σ(X)
print("2. σ-algebra Generated by X, σ(X):")
print("   Since X(t,p) = t, σ(X) consists of sets of the form A × ℝ, where A is a Borel set in ℝ.")
print("   These are cylindrical sets, where the pressure component is unrestricted.\n")

# 3. σ-algebra generated by Y, σ(Y)
print("3. σ-algebra Generated by Y, σ(Y):")
print("   Since Y(t,p) = (t,p), σ(Y) is the Borel σ-algebra on ℝ², ℬ(ℝ²).\n")

# 4. Show that σ(X) ⊆ σ(Y) and explain why Y contains more information than X.
print("4. Information Comparison (σ(X) ⊆ σ(Y)):")
print("   Any set in σ(X) is of the form A × ℝ, which is a subset of ℝ² and thus belongs to ℬ(ℝ²) = σ(Y). Hence, σ(X) ⊆ σ(Y).")
print("   Y contains more information than X because Y records both temperature deviation (t) and pressure deviation (p), while X only records temperature deviation (t).")
print("   Knowing Y means knowing both t and p, whereas knowing X only means knowing t, leaving p unknown.\n")

# 5. Marginal density of X (T)
print("5. Marginal Density of X:")
print("   X ~ N(0, 1). PDF: f_X(x) = (1/sqrt(2*pi)) * exp(-x^2 / 2)\n")

# 6. Joint Density of Y (Full sensor state)
print("6. Joint Density of Y (Full sensor state):")
print("   f_Y(t, p) = [1 / (2*pi*1*2)] * exp(-0.5 * (t^2/1 + p^2/4))\n")

# 7. Compute probabilities
print("7. Compute the following probabilities:")
# P(X <= 0) for X ~ N(0,1)
p_X_le_0 = norm.cdf(0, loc=0, scale=1)
print(f"   P(X ≤ 0) = {p_X_le_0:.4f}")

# P(X > 1) for X ~ N(0,1)
p_X_gt_1 = 1 - norm.cdf(1, loc=0, scale=1)
print(f"   P(X > 1) = {p_X_gt_1:.4f}")

# P(Y ∈ (-∞,0] × ℝ) is equivalent to P(T ≤ 0)
p_Y_interval1 = p_X_le_0
print(f"   P(Y ∈ (-∞,0] × ℝ) = P(T ≤ 0) = {p_Y_interval1:.4f}")

# P(Y ∈ [-1,1] × [-2,2]) is P(-1 ≤ T ≤ 1) * P(-2 ≤ P ≤ 2) due to independence
p_T_interval = norm.cdf(1, loc=0, scale=1) - norm.cdf(-1, loc=0, scale=1)
p_P_interval = norm.cdf(2, loc=0, scale=2) - norm.cdf(-2, loc=0, scale=2) # P ~ N(0,4) => std dev is 2
p_Y_interval2 = p_T_interval * p_P_interval
print(f"   P(Y ∈ [-1,1] × [-2,2]) = {p_Y_interval2:.4f}\n")

# 8. Conditional distribution of X given Y=(t,p)
print("8. Conditional Distribution of X given Y=(t,p):")
print("   Since X(t,p) = t, if Y=(t,p) is known, then X is known to be t with certainty.")
print("   The conditional distribution of X given Y=(t,p) is a degenerate distribution (a Dirac delta function) at t.\n")

# 9. Conditional distribution of Y given X=t
print("9. Conditional Distribution of Y given X=t:")
print("   Given X=t (i.e., T=t), and since T and P are independent, knowing T tells us nothing about P.")
print("   The conditional distribution of Y=(T,P) given X=t is (t, P'), where P' follows its marginal distribution N(0,4).\n")

# 10. Explain what is known when one conditions on X and what additional information becomes available when one conditions on Y.
print("10. Interpretation of Conditioning on X vs. Y:")
print("    - When conditioning on X=t, the exact temperature deviation t is known. However, the pressure deviation remains unknown, and its distribution is still N(0,4).")
print("    - When conditioning on Y=(t,p), both the exact temperature deviation t and the exact pressure deviation p are known.")
print("      No uncertainty remains about either variable.\n")

# 11. Show that X = π1 ◦ Y, where π1(t,p)=t, and explain why this implies that X is a measurable function of Y.
print("11. X as a measurable function of Y:")
print("    Given Y(t,p)=(t,p) and π1(t,p)=t.")
print("    Then (π1 ◦ Y)(t,p) = π1(Y(t,p)) = π1(t,p) = t.")
print("    Also, X(t,p)=t. Thus, X = π1 ◦ Y.")
print("    This implies X is a measurable function of Y because π1 is a projection map, which is a continuous (and thus measurable) function from ℝ² to ℝ.")
print("    The composition of measurable functions is measurable, confirming σ(X) ⊆ σ(Y).\n")

# 12. Visualization
print("12. Python-based Visualization of the Model:")
# Setup for plotting
x = np.linspace(-3, 3, 100)
y = np.linspace(-6, 6, 100)
X_grid, P_grid = np.meshgrid(x, y)
pos = np.dstack((X_grid, P_grid))
rv = multivariate_normal(mean, cov)
Z = rv.pdf(pos)

plt.figure(figsize=(12, 5))

# Joint Density Contour
plt.subplot(1, 2, 1)
plt.contourf(X_grid, P_grid, Z, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Joint Density $f_{T,P}(t, p)$')
plt.xlabel('Temperature (T)')
plt.ylabel('Pressure (P)')

# Marginal Density of X
plt.subplot(1, 2, 2)
x_axis = np.linspace(-4, 4, 100)
plt.plot(x_axis, norm.pdf(x_axis, loc=0, scale=1), color='blue', lw=2)
plt.fill_between(x_axis, norm.pdf(x_axis, loc=0, scale=1), alpha=0.3)
plt.title('Marginal Density of X (Temperature)')
plt.xlabel('x')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

print("    *The contour plot shows the joint density of Y=(T,P), illustrating how temperature and pressure deviations are distributed.")
print("    *The line plot shows the marginal density of X (Temperature), which is a standard normal distribution.")
print("    *Illustrating that σ(X) is coarser than σ(Y): Conceptually, σ(Y) allows us to distinguish between any two distinct points (t,p) and (t',p') in ℝ². σ(X) only allows distinguishing points if their temperature components (t and t') are different. If t=t', then regardless of p or p', σ(X) cannot distinguish them. This means σ(X) partitions ℝ² into 'vertical strips' (constant t), while σ(Y) can isolate individual points, hence σ(X) is coarser.\n")



# 13. Differential Entropy h(X)
# h(Normal) = 0.5 * ln(2 * pi * e * variance)
def diff_entropy_normal(var):
    return 0.5 * np.log(2 * np.pi * np.e * var)

h_X = diff_entropy_normal(1) # Variance of T is 1
print("13. Differential Entropy h(X):")
print(f"    h(X) = {h_X:.4f} nats\n")

# 14. Differential Entropy h(Y)
# For multivariate normal: h(Y) = 0.5 * ln((2*pi*e)^k * |det(Cov)|)
# k is dimension (2), det(Cov) is 1 * 4 = 4
h_Y = 0.5 * np.log(np.linalg.det(2 * np.pi * np.e * np.array(cov)))
print("14. Differential Entropy h(Y):")
print(f"    h(Y) = {h_Y:.4f} nats\n")

# 15. Conditional Differential Entropy h(Y|X)
# h(Y|X) = h((T,P)|T) = h(P|T). Since T and P are independent, h(P|T) = h(P).
# P ~ N(0,4), so variance of P is 4.
h_Y_given_X = diff_entropy_normal(4)
print("15. Conditional Differential Entropy h(Y|X):")
print(f"    h(Y|X) = h(P) = {h_Y_given_X:.4f} nats")
print("    Interpretation: This is the remaining uncertainty in the pressure sensor given temperature is known, which, due to independence, is simply the entropy of the pressure itself.\n")

# 16. Discuss what happens to h(X|Y) and why the continuous case differs from the discrete case.
print("16. Discussion of h(X|Y):")
print("    In this model, X(t,p) = t, and Y(t,p) = (t,p). So, if Y=(t,p) is known, then X is completely determined to be t.")
print("    Therefore, there is no uncertainty about X given Y. In a discrete setting (like Q9), this would lead to H(X|Y) = 0.")
print("    However, for continuous random variables, the differential entropy of a degenerate distribution (a point mass at t) is $-\infty$.")
print("    Thus, h(X|Y) = $-\infty$.")
print("    This difference arises because discrete entropy uses logarithms of probabilities, where a point event can have probability 1 (log(1)=0). Differential entropy uses logarithms of probability *densities*. The probability density function for a point mass is a Dirac delta function, and its integral (which would be used in the entropy calculation) leads to $-\infty$.")
print("    The intuitive meaning of 'no uncertainty' is the same, but its mathematical representation differs between discrete (0) and continuous ($-\infty$) entropy.\n")
```

#Q11) Conditional Expectation

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer’s risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets ${ \omega_1,\omega_2}$ and ${\omega_3,\omega_4}$.

4. Compute $E[X\mid Y]$ explicitly.

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Compute $E[X]$ and compare it with $E[X\mid Y]$.

7. Interpret $E[X\mid Y]$ in practical terms. What does it tell the bank?

8. Let

$$
\mathfrak G=\mathfrak P(\Omega).
$$

Compute $E[X\mid \mathfrak G]$ and compare it with $E[X\mid Y]$.

9. Explain how this example shows that conditional expectation is the “best estimate of $X$ using only the information contained in $Y$.”

```python
import numpy as np

# Outcome probabilities
# w1: Low, Legitimate; w2: Low, Fraud; w3: High, Legitimate; w4: High, Fraud
probs = {'w1': 0.50, 'w2': 0.10, 'w3': 0.20, 'w4': 0.20}

print("--- Q11) Conditional Expectation: Fraud Detection ---\n")

# 1. Write down the probability space (Ω, F, P)
print("1. Probability Space (Ω, F, P):")
omega_space = set(probs.keys())
print(f"   Ω (Sample Space) = {omega_space}")
print(f"   F (Sigma-algebra) = P(Ω) (the power set of Ω, as Ω is finite)")
print(f"   P (Probability Measure): Defined by P({{ω_i}}) as given:")
for w, p in probs.items():
    print(f"     P({w}) = {p}")
print("\n")

# Define the random variable X (fraud indicator)
def X(omega):
    if omega in ['w2', 'w4']:
        return 1  # Fraudulent
    else:
        return 0  # Legitimate

# Define the random variable Y (customer risk class)
def Y(omega):
    if omega in ['w1', 'w2']:
        return 0  # Low-risk
    else:
        return 1  # High-risk

# 2. Compute the σ-algebra generated by Y, namely σ(Y)
print("2. Compute the σ-algebra generated by Y, namely σ(Y):")
# Y=0 corresponds to {w1, w2}
# Y=1 corresponds to {w3, w4}
set_Y0 = {'w1', 'w2'}
set_Y1 = {'w3', 'w4'}

sigma_Y = [
    set(),                  # Empty set
    set_Y0,                 # Y=0
    set_Y1,                 # Y=1
    set_Y0.union(set_Y1)    # Ω
]
print(f"   σ(Y) = {sigma_Y}\n")

# 3. Explain why E[X|Y] must be σ(Y)-measurable, and therefore must be constant on the sets {ω1,ω2} and {ω3,ω4}.
print("3. Explanation of E[X|Y] σ(Y)-measurable:")
print("   The conditional expectation E[X|Y] is defined as a random variable that is σ(Y)-measurable.")
print("   This means that its value must be constant on any set in σ(Y)'s partition of Ω.")
print("   The partition generated by Y is {{w1,w2}, {w3,w4}}. Therefore, E[X|Y] will take one constant value when Y=0 (on {w1,w2}) and another constant value when Y=1 (on {w3,w4}).\n")

# 4. Compute E[X|Y] explicitly
print("4. Compute E[X|Y] explicitly:")
# Calculate P(Y=0) and P(Y=1)
p_Y0 = probs['w1'] + probs['w2']  # P(Y=0) = P({w1, w2}) = 0.50 + 0.10 = 0.60
p_Y1 = probs['w3'] + probs['w4']  # P(Y=1) = P({w3, w4}) = 0.20 + 0.20 = 0.40

# E[X | Y=0] = P(X=1 | Y=0) = P(X=1, Y=0) / P(Y=0) = P({w2}) / P({w1, w2})
e_x_given_y0 = probs['w2'] / p_Y0

# E[X | Y=1] = P(X=1 | Y=1) = P(X=1, Y=1) / P(Y=1) = P({w4}) / P({w3, w4})
e_x_given_y1 = probs['w4'] / p_Y1

print(f"   E[X|Y=0] = P(X=1|Y=0) = P({{w2}}) / P({{w1,w2}}) = {probs['w2']} / {p_Y0} = {e_x_given_y0:.4f}")
print(f"   E[X|Y=1] = P(X=1|Y=1) = P({{w4}}) / P({{w3,w4}}) = {probs['w4']} / {p_Y1} = {e_x_given_y1:.4f}\n")

# 5. Verify the defining property of conditional expectation
print("5. Verify the defining property of conditional expectation:")
# For A in σ(Y), we need ∫_A E[X|Y] dP = ∫_A X dP
# This means: E[E[X|Y] * I_A] = E[X * I_A]

# Case 1: A = {w1, w2} (where Y=0)
# LHS: E[E[X|Y] * I_{w1,w2}] = E[X|Y=0] * P({w1,w2}) = e_x_given_y0 * p_Y0
# RHS: E[X * I_{w1,w2}] = X(w1)*P(w1) + X(w2)*P(w2) = 0*0.50 + 1*0.10
lhs1 = e_x_given_y0 * p_Y0
rhs1 = X('w1') * probs['w1'] + X('w2') * probs['w2']
print(f"   For A = {{w1, w2}} (Y=0):\n     LHS (E[E[X|Y]*I_A]) = {e_x_given_y0:.4f} * {p_Y0:.2f} = {lhs1:.4f}")
print(f"     RHS (E[X*I_A])      = {X('w1')}*{probs['w1']:.2f} + {X('w2')}*{probs['w2']:.2f} = {rhs1:.4f}")
print(f"     LHS == RHS: {np.isclose(lhs1, rhs1)}\n")

# Case 2: A = {w3, w4} (where Y=1)
# LHS: E[E[X|Y] * I_{w3,w4}] = E[X|Y=1] * P({w3,w4}) = e_x_given_y1 * p_Y1
# RHS: E[X * I_{w3,w4}] = X(w3)*P(w3) + X(w4)*P(w4) = 0*0.20 + 1*0.20
lhs2 = e_x_given_y1 * p_Y1
rhs2 = X('w3') * probs['w3'] + X('w4') * probs['w4']
print(f"   For A = {{w3, w4}} (Y=1):\n     LHS (E[E[X|Y]*I_A]) = {e_x_given_y1:.4f} * {p_Y1:.2f} = {lhs2:.4f}")
print(f"     RHS (E[X*I_A])      = {X('w3')}*{probs['w3']:.2f} + {X('w4')}*{probs['w4']:.2f} = {rhs2:.4f}")
print(f"     LHS == RHS: {np.isclose(lhs2, rhs2)}\n")

# Case 3: A = Ω
lhs3 = lhs1 + lhs2
rhs3 = rhs1 + rhs2
print(f"   For A = Ω:\n     LHS (E[E[X|Y]*I_Ω]) = {lhs3:.4f}")
print(f"     RHS (E[X*I_Ω])      = {rhs3:.4f}")
print(f"     LHS == RHS: {np.isclose(lhs3, rhs3)}\n")

# 6. Compute E[X] and compare it with E[X|Y]
print("6. Compute E[X] and compare it with E[X|Y]:")
E_X = sum(X(w) * p for w, p in probs.items())
print(f"   E[X] = {E_X:.4f} (Overall probability of a transaction being fraudulent)\n")

# Compare with E[X|Y] using the Law of Total Expectation
E_E_X_given_Y = e_x_given_y0 * p_Y0 + e_x_given_y1 * p_Y1
print(f"   E[E[X|Y]] = E[X|Y=0]P(Y=0) + E[X|Y=1]P(Y=1)")
print(f"             = ({e_x_given_y0:.4f} * {p_Y0:.2f}) + ({e_x_given_y1:.4f} * {p_Y1:.2f}) = {E_E_X_given_Y:.4f}")
print(f"   We see that E[E[X|Y]] = E[X], which is {np.isclose(E_X, E_E_X_given_Y)}.\n")

# 7. Interpret E[X|Y] in practical terms. What does it tell the bank?
print("7. Interpretation of E[X|Y] in practical terms:")
print("   E[X|Y] represents the bank's *updated belief* or *best estimate* of the probability of fraud, given the information about the customer's risk class (Y).")
print(f"   - If a customer is identified as 'Low-risk' (Y=0), the estimated probability of fraud for their transaction is E[X|Y=0] = {e_x_given_y0:.2%}.")
print(f"   - If a customer is identified as 'High-risk' (Y=1), the estimated probability of fraud for their transaction is E[X|Y=1] = {e_x_given_y1:.2%}.")
print(f"   This tells the bank that low-risk customers have a much lower fraud rate ({e_x_given_y0:.2%}) than high-risk customers ({e_x_given_y1:.2%}), compared to the overall average fraud rate of {E_X:.2%}. This information can be used to tailor fraud detection strategies, such as applying more scrutiny to transactions from high-risk customers.\n")

# 8. Compute E[X|G] and compare it with E[X|Y]
print("8. Compute E[X|G] where G = P(Ω) and compare with E[X|Y]:")
print("   If G = P(Ω), then G contains all possible events, including the elementary events {ω1}, {ω2}, {ω3}, {ω4}.")
print("   In this case, conditioning on G means we know the exact outcome ω with certainty.")
print("   Therefore, E[X|G] is simply X itself. For each ω_i, E[X|G](ω_i) = X(ω_i).")
print(f"   - E[X|G](ω1) = X(ω1) = {X('w1')}")
print(f"   - E[X|G](ω2) = X(ω2) = {X('w2')}")
print(f"   - E[X|G](ω3) = X(ω3) = {X('w3')}")
print(f"   - E[X|G](ω4) = X(ω4) = {X('w4')}\n")
print("   Comparing with E[X|Y]: E[X|Y] provides an average expectation based on the coarser information provided by Y. E[X|G] provides the exact value of X because G contains maximal information.\n")

# 9. Explain how this example shows that conditional expectation is the “best estimate of X using only the information contained in Y.”
print("9. Explanation: Conditional expectation as 'best estimate':")
print("   The conditional expectation E[X|Y] acts as the 'best estimate' of X (the fraud indicator) given only the information available in Y (the customer's risk class).")
print("   - It's the 'best' estimate in the sense that it minimizes the mean squared error (MSE) among all σ(Y)-measurable random variables. That is, E[(X - E[X|Y])²] ≤ E[(X - Z)²] for any other σ(Y)-measurable random variable Z.")
print("   - The example illustrates this: without knowing Y, the bank's best guess for fraud is E[X] = {E_X:.2%}. However, once Y is known, the estimate improves to {e_x_given_y0:.2%} for low-risk customers and {e_x_given_y1:.2%} for high-risk customers.")
print("   - This shows that E[X|Y] effectively extracts all relevant information about X that is contained within Y, providing a more refined and accurate prediction than the overall expectation E[X], but not as precise as knowing the full outcome (X itself).")
```
